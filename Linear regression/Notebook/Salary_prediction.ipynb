```python
# importing modules and packages

import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn.linear_model import LinearRegression
from word2number import w2n
```

```python
path = '/content/drive/MyDrive/Colab Notebooks/Machine learning/dataset/salary.csv'
df = pd.read_csv(path)
df
```

Preapre data and feature engineering


```python
# handling missing values

df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(df['test_score(out of 10)'].mean()).round(1)
df['experience'] = df['experience'].fillna('zero')
df
```

```python
# convert the experience column into numbers

df['experience'] = df['experience'].apply(w2n.word_to_num)
```

```python
df
```

```python
X = df.iloc[:,:-1]
Y = df['salary($)']
```

```python
reg = LinearRegression()
reg.fit(X,Y)
reg.score(X,Y)
```

```python
# predict on unseen data
# 2 yrs exp.9 test score,6 interview score

reg.predict([[2,9,6]]).round(2)
```
